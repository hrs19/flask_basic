```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv("dataset/Iris.data") 
# from sklearn import datasets
# iris = datasets.load_iris()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
X = iris.drop('class',axis=1)
y = iris['class']
y = encode.fit_transform(y)
print(X.head())
print(y)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.svm import SVC
from sklearn import metrics
SVC_model = SVC()
SVC_model.fit(X_train,y_train)
SVC_predict = SVC_model.predict(X_test)
print('The accuracy of the SVM is:', metrics.accuracy_score(SVC_predict,y_test))
```

```python
from sklearn.linear_model import LogisticRegression 
LR_model = LogisticRegression()
LR_model.fit(X_train,y_train)
LR_prediction=LR_model.predict(X_test)
print('The accuracy of the Logistic Regression is',metrics.accuracy_score(LR_prediction,y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
DT_model=DecisionTreeClassifier()
DT_model.fit(X_train,y_train)
DT_prediction=DT_model.predict(X_test)
print('The accuracy of the Decision Tree is',metrics.accuracy_score(DT_prediction,y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier

KNN_model=KNeighborsClassifier(n_neighbors=3) #this examines 3 neighbours for putting the new data into a class
KNN_model.fit(X_train,y_train)
KNN_prediction=KNN_model.predict(X_test)
print('The accuracy of the KNN is',metrics.accuracy_score(KNN_prediction,y_test))
```
